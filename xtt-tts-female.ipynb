```python
import torch
import torchaudio
import os
from TTS.api import TTS
from IPython.display import Audio

# 1. FORCE DELETE the corrupted file if it exists
if os.path.exists("reference.wav"):
    os.remove("reference.wav")

# 2. Download a GUARANTEED valid wav file (LibriSpeech sample)
# This uses a direct link to a real audio file
!wget -q -O reference.wav https://www.pacdv.com/sounds/voices/maybe-next-time.wav

# 3. Setup paths
model_path = "./kinyarwanda_health_model"
speaker_wav = "reference.wav"

# 4. Load the model (Only if not already loaded)
try:
    if 'tts' not in locals():
        tts = TTS(model_path=model_path, config_path=os.path.join(model_path, "config.json")).to("cuda")
    model = tts.synthesizer.tts_model
except Exception as e:
    print(f"Model Load Error: {e}")

# 5. Manual Inference
def speak_kinyarwanda(text_to_speak, output_name="kinyarwanda_final.wav"):
    print("Processing audio...")
    # Get voice characteristics
    gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=speaker_wav)

    # Run inference using "en" as the proxy language
    out = model.inference(
        text_to_speak,
        "en",
        gpt_cond_latent,
        speaker_embedding,
        temperature=0.7,
        repetition_penalty=5.0
    )

    wav = torch.tensor(out['wav']).unsqueeze(0)
    torchaudio.save(output_name, wav, 24000)
    return output_name

# 6. Run it
text = """
  Ikoranabuhanga ni umusingi w'iterambere ryiza
"""
try:
    file = speak_kinyarwanda(text)
    print(f"✅ Success! Created {file}")
    display(Audio(file, autoplay=True))
except Exception as e:
    print(f"❌ Generation Error: {e}")
```
